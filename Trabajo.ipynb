```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Colocar la tabla completa

```python
Metropolitana_sur = pd.read_excel('Metropolitana_Sur.xlsx')

Metropolitana_sur
```

```python
Genero = Metropolitana_sur['GENERO']
Rango_Edad = Metropolitana_sur['RANGO EDAD']
Tipo_Institucion = Metropolitana_sur['TIPO DE INSTITUCION']
Area_Conocimiento = Metropolitana_sur['AREA CONOCIMIENTO']
Comuna_Sede = Metropolitana_sur['COMUNA SEDE']
Nivel_Carrera = Metropolitana_sur['NIVEL CARRERA']
```

Grafico de barras                                         
-Genero                                                      
-Tipo de Institucion                                                        
-Area de Conocimiento

Primero el grafico de barra en Genero

```python
genero_leng = Genero.value_counts()
ind_genero = genero_leng.index
val_genero = genero_leng.values

fig, ax = plt.subplots(figsize=(6,4))

columna_g = ax.bar(ind_genero, val_genero, zorder=2)

ax.bar_label(columna_g)
#diseño

ax.grid(zorder=1)
ax.set_title('Genero de los estudiantes')
ax.set_ylabel('Cantidad de personas')
ax.set_ylim(0,80000)

plt.xticks(rotation=0, ha='right')
plt.tight_layout()

plt.show()
```

Grafico de Tipo de Institucion

```python
institucion_leng = Tipo_Institucion.value_counts()

ind_institucion = institucion_leng.index
val_institucion = institucion_leng.values

fig, ax = plt.subplots(figsize=(6,4))

columna_i = ax.bar(ind_institucion, val_institucion, zorder=2)
ax.bar_label(columna_i)

#diseño

ax.grid(zorder=1)
ax.set_title('Tipos de Instituciones')
ax.set_ylabel('Cantidad de personas')
ax.set_ylim(0,60000)

plt.xticks(rotation=20, ha='right')
plt.tight_layout()

plt.show()
```

Grafico de Areas de Conocimientos

```python
areas_leng = Area_Conocimiento.value_counts()
ind_areas = areas_leng.index
val_areas = areas_leng.values

fig, ax = plt.subplots(figsize=(6,4))

columna_a = ax.bar(ind_areas, val_areas, zorder=2)
ax.bar_label(columna_a)

#diseño
ax.grid(zorder=1)
ax.set_title('Areas de Conocimiento')
ax.set_ylabel('Cantidad de personas')
ax.set_ylim(0,60000)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

GRAFICO CIRCULAR                                                   
-Region Sede                                                             
-Nivel Carrera

Grafico Region

```python
comuna_leng = Comuna_Sede.value_counts()
ind_comuna = comuna_leng.index
val_comuna = comuna_leng.values

fig, ax = plt.subplots(figsize=(6,6))

wedges, texts, autotexts = ax.pie(
    val_comuna,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False
)

for t in autotexts:
    t.set_visible(False)


ax.legend(wedges,
          [f"{name} ({pct.get_text()})" for name, pct in zip(ind_comuna, autotexts)],
          title='Comunas',
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1),
          )

ax.set_title('Comuna Sede')
plt.show()
```

Grafico Nivel de Carrera

```python
nivel_leng = Nivel_Carrera.value_counts()
ind_nivel = nivel_leng.index
val_nivel = nivel_leng.values

fig, ax = plt.subplots(figsize=(5,5))

wedges, texts, autotexts = ax.pie(
    val_nivel,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False
)

for t in autotexts:
    t.set_visible(False)


ax.legend(wedges,
          [f"{name} ({pct.get_text()})" for name, pct in zip(ind_nivel, autotexts)],
          title='Nivel Carrera',
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1),
          )

ax.set_title('Niveles de Carrera')
plt.show()

```

Medidas de Dispersion en las Edades de los Estudiantes

```python
Edades = Metropolitana_sur['EDAD']

Rango = Edades.max() - Edades.min()
Desviacion = Edades.std()
var = Edades.var()
cv = (Desviacion / Edades.mean()) * 100

print(f'Rango de edades {Rango}')
print(f'desviacion estandar {Desviacion:.2f}')
print(f'varianza {var:.2f}')
print(f'coeficiente de variacion {cv:.2f}%')
```

Medidas de Dispersion en el plan estudiantil de los estudiantes

```python
Plan_E = Metropolitana_sur['DURACION PLAN DE ESTUDIO (SEMESTRES)']

Rango = Plan_E.max() - Plan_E.min()
Desviacion = Plan_E.std()
var = Plan_E.var()
cv = (Desviacion / Plan_E.mean()) * 100

print(f'Rango de Duracion Plan de Estudio {Rango}')
print(f'desviacion estandar del plan de estudios {Desviacion:.2f}')
print(f'varianza del plan de estudios {var:.2f}')
print(f'coeficiente de variacion del plan de estudios {cv:.2f}%')
```

Duracion Proceso de Titulacion

```python
Proceso_e = Metropolitana_sur['DURACION PROCESO TITULACION (SEMESTRES)']
Rango = Proceso_e.max() - Proceso_e.min()
Desviacion = Proceso_e.std()
var = Proceso_e.var()
cv = (Desviacion / Proceso_e.mean()) * 100

print(f'Rango de Duracion Proceso de Titulacion {Rango}')
print(f'desviacion estandar del proceso de titulacion {Desviacion:.2f}')
print(f'varianza del proceso de titulacion {var:.2f}')
print(f'coeficiente de variacion del proceso de titulacion {cv:.2f}%')
```

Duracion total de la carrera

```python
Duracion_T = Metropolitana_sur['DURACION TOTAL CARRERA (SEMESTRES)']
Rango = Duracion_T.max() - Duracion_T.min()
Desviacion = Duracion_T.std()
var = Duracion_T.var()
cv = (Desviacion / Duracion_T.mean()) * 100

print(f'Rango de Duracion Total de la Carrera {Rango}')
print(f'desviacion estandar de la duracion total de la carrera {Desviacion:.2f}')
print(f'varianza de la duracion total de la carrera {var:.2f}')
print(f'coeficiente de variacion de la duracion total de la carrera {cv:.2f}%')
```

Valor de la matricula

```python
Valor_M = Metropolitana_sur['VALOR MATRICULA (PESOS)']
Rango = Valor_M.max() - Valor_M.min()
Desviacion = Valor_M.std()
var = Valor_M.var()
cv = (Desviacion / Valor_M.mean()) * 100

print(f'Rango del Valor de la Matricula ${Rango}')
print(f'desviacion estandar del valor de la matricula ${Desviacion:.0f}')
print(f'varianza del valor de la matricula ${var:.0f}')
print(f'coeficiente de variacion del valor de la matricula {cv:.2f}%')
```
